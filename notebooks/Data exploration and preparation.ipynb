```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats
import warnings
from datetime import datetime
warnings.filterwarnings('ignore')

# ËÆæÁΩÆÂ≠ó‰ΩìÂíåÊ†∑Âºè
plt.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = [10, 6]
plt.style.use('seaborn-v0_8-whitegrid')

print("="*80)
print("Ëà™Á©∫ÂÖ¨Âè∏‰πòÂÆ¢Êª°ÊÑèÂ∫¶Êï∞ÊçÆÂàÜÊûê - Student ID: 14645473")
print("="*80)

# ÂàõÂª∫Â∏¶Êó∂Èó¥Êà≥ÁöÑËæìÂá∫Êñá‰ª∂Â§π
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
output_folder = f'/Users/suzydon/Desktop/assignment_analysis_{timestamp}'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"‚úÖ Created analysis folder: {output_folder}")

# Âä†ËΩΩÊï∞ÊçÆÈõÜ
try:
    file_path = '/Users/suzydon/Desktop/32130_AT2_14645473.csv'
    df = pd.read_csv(file_path)
    print(df.info())
    print(df.describe(include='all'))
    print(f"‚úÖ Dataset loaded successfully from: {file_path}")
    print(f"üìä Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
except FileNotFoundError:
    print(f"‚ùå Error: File not found at {file_path}")
    print("Please check if the file exists and the path is correct.")
    exit()

# ÂàõÂª∫ExcelÊñá‰ª∂Ë∑ØÂæÑ
excel_file_path = f'/Users/suzydon/Desktop/fda_a2_14645473_{timestamp}.xlsx'

print("\n" + "="*80)
print("TASK 1A: INITIAL DATA EXPLORATION")
print("="*80)

# Task 1A.1: ËØÜÂà´Â±ûÊÄßÁ±ªÂûã
print("\nüìã Task 1A.1: Identifying Attribute Types")
print("-" * 50)

print("\nüîç Dataset Information:")
df.info()

print("\nüìä Data Types Summary:")
for dtype in df.dtypes.unique():
    cols = df.select_dtypes(include=[dtype]).columns.tolist()
    print(f"  {dtype}: {len(cols)} attributes - {cols}")

# ËØ¶ÁªÜÁöÑÁº∫Â§±ÂÄºÊ£ÄÊü•
print("\nüï≥Ô∏è Missing Values Analysis:")
missing_values = df.isnull().sum()
total_missing = missing_values.sum()

if total_missing == 0:
    print("  ‚úÖ No missing values found!")
else:
    print(f"  ‚ö†Ô∏è Total missing values: {total_missing}")
    print("  Missing values by column:")
    for col, count in missing_values[missing_values > 0].items():
        percentage = (count / len(df)) * 100
        print(f"    {col}: {count} ({percentage:.2f}%)")
    
    # ÁâπÂà´ÂÖ≥Ê≥®Âà∞ËææÂª∂ËØØÁöÑÁº∫Â§±ÂÄº
    if 'Arrival Delay in Minutes' in missing_values.index and missing_values['Arrival Delay in Minutes'] > 0:
        arrival_missing = missing_values['Arrival Delay in Minutes']
        print(f"\n  üìä Arrival Delay Detailed Analysis:")
        print(f"    Missing count: {arrival_missing}")
        print(f"    Missing percentage: {(arrival_missing/len(df)*100):.2f}%")
        print(f"    Valid records: {len(df) - arrival_missing}")
        print(f"    Data completeness: {((len(df) - arrival_missing)/len(df)*100):.2f}%")

# Ê£ÄÊü•ÂàÜÁ±ªÂèòÈáèÁöÑÂîØ‰∏ÄÂÄº
categorical_columns = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']
print("\nüè∑Ô∏è Categorical Attributes Analysis:")
for col in categorical_columns:
    if col in df.columns:
        unique_vals = df[col].unique()
        print(f"  '{col}': {len(unique_vals)} unique values - {unique_vals}")

# Task 1A.2: Ê±áÊÄªÂ±ûÊÄßÁâπÊÄß
print("\nüìà Task 1A.2: Summarizing Properties")
print("-" * 50)

# ËØÜÂà´Êï∞ÂÄºÂíåÂàÜÁ±ªÂ±ûÊÄß
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print(f"\nüî¢ Numerical attributes ({len(numeric_cols)}): {numeric_cols}")
print(f"üè∑Ô∏è Categorical attributes ({len(categorical_cols)}): {categorical_cols}")

# Êï∞ÂÄºÂ±ûÊÄßÊ±áÊÄªÁªüËÆ°
print("\nüìä Summary Statistics for Numerical Attributes:")
numeric_summary = df[numeric_cols].describe()
print(numeric_summary)

# ‰∏∫Êï∞ÂÄºÂ±ûÊÄßÁîüÊàêËØ¶ÁªÜÂèØËßÜÂåñ
print(f"\nüé® Generating visualizations for numerical attributes...")
for i, col in enumerate(numeric_cols):
    clean_col = col.replace('/', '_').replace(' ', '_').replace('-', '_').replace('(', '').replace(')', '')
    
    print(f"  üìä Analyzing: {col} ({i+1}/{len(numeric_cols)})")
    
    # Ê£ÄÊü•Áº∫Â§±ÂÄº
    missing_count = df[col].isnull().sum()
    valid_data = df[col].dropna()
    
    if len(valid_data) == 0:
        print(f"    ‚ö†Ô∏è Warning: No valid data for {col}")
        continue
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    title_suffix = f" (Valid: {len(valid_data)}, Missing: {missing_count})" if missing_count > 0 else f" (Total: {len(valid_data)})"
    fig.suptitle(f'Detailed Analysis of {col}{title_suffix}', fontsize=16, fontweight='bold')
    
    # 1. Áõ¥ÊñπÂõæÂíåÂØÜÂ∫¶Âõæ
    sns.histplot(valid_data, kde=True, ax=axes[0,0], stat='density')
    axes[0,0].set_title(f'Distribution of {col}')
    axes[0,0].set_xlabel(col)
    axes[0,0].set_ylabel('Density')
    
    # 2. ÁÆ±Á∫øÂõæ
    sns.boxplot(x=valid_data, ax=axes[0,1])
    axes[0,1].set_title(f'Box Plot of {col}')
    axes[0,1].set_xlabel(col)
    
    # 3. Q-QÂõæÊ£ÄÈ™åÊ≠£ÊÄÅÊÄß
    try:
        stats.probplot(valid_data, dist="norm", plot=axes[1,0])
        axes[1,0].set_title(f'Q-Q Plot of {col}')
    except:
        axes[1,0].text(0.5, 0.5, 'Q-Q Plot unavailable', ha='center', va='center', transform=axes[1,0].transAxes)
        axes[1,0].set_title(f'Q-Q Plot of {col}')
    
    # 4. Á¥ØÁßØÂàÜÂ∏ÉÂõæ
    sorted_data = np.sort(valid_data)
    cumulative = np.arange(1, len(sorted_data) + 1) / len(sorted_data)
    axes[1,1].plot(sorted_data, cumulative, marker='.', linestyle='none')
    axes[1,1].set_title(f'Cumulative Distribution of {col}')
    axes[1,1].set_xlabel(col)
    axes[1,1].set_ylabel('Cumulative Probability')
    axes[1,1].grid(True)
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'{clean_col}_detailed_analysis.png'), 
                dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

# ÂàÜÁ±ªÂ±ûÊÄßÂàÜÊûê
print(f"\nüè∑Ô∏è Analyzing categorical attributes...")
for i, col in enumerate(categorical_cols):
    clean_col = col.replace('/', '_').replace(' ', '_').replace('-', '_').replace('(', '').replace(')', '')
    
    print(f"  üìä Analyzing: {col} ({i+1}/{len(categorical_cols)})")
    
    value_counts = df[col].value_counts()
    print(f"    Frequency distribution:")
    for val, count in value_counts.items():
        percentage = (count / len(df)) * 100
        print(f"      {val}: {count:,} ({percentage:.1f}%)")
    
    # ÂàõÂª∫È¢ëÁéáÂõæ
    plt.figure(figsize=(12, 8))
    colors = plt.cm.Set3(np.linspace(0, 1, len(value_counts)))
    bars = plt.barh(range(len(value_counts)), value_counts.values, color=colors)
    plt.yticks(range(len(value_counts)), value_counts.index)
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.title(f'Frequency Distribution of {col}')
    
    # Ê∑ªÂä†Êï∞ÂÄºÂíåÁôæÂàÜÊØîÊ†áÁ≠æ
    total = len(df)
    for i, (bar, count) in enumerate(zip(bars, value_counts.values)):
        percentage = (count / total) * 100
        plt.text(bar.get_width() + total*0.01, bar.get_y() + bar.get_height()/2, 
                f'{count:,} ({percentage:.1f}%)', 
                va='center', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, f'{clean_col}_frequency_distribution.png'), 
                dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

# Task 1A.3: Â§öÂ±ûÊÄßÂÖ≥Á≥ªÊé¢Á¥¢
print("\nüîó Task 1A.3: Exploring Multiple Attributes Relationships")
print("-" * 50)

# Áõ∏ÂÖ≥ÊÄßÁü©Èòµ - Âè™ÂØπÊï∞ÂÄºÂàó‰∏îÁº∫Â§±ÂÄºËæÉÂ∞ëÁöÑÂàóËÆ°ÁÆó
print("üìä Calculating correlation matrix...")
numeric_for_corr = []
for col in numeric_cols:
    missing_pct = df[col].isnull().sum() / len(df)
    if missing_pct < 0.5:  # Áº∫Â§±ÂÄºÂ∞ë‰∫é50%ÁöÑÂàó
        numeric_for_corr.append(col)

if len(numeric_for_corr) > 1:
    correlation_matrix = df[numeric_for_corr].corr()
    
    plt.figure(figsize=(14, 10))
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, 
                mask=mask, 
                annot=True, 
                cmap='RdBu_r', 
                center=0, 
                square=True, 
                fmt='.2f',
                cbar_kws={"shrink": .8})
    plt.title('Correlation Matrix of Numerical Attributes', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'correlation_matrix.png'), 
                dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
    
    # Âº∫Áõ∏ÂÖ≥ÊÄßËØÜÂà´
    print("\nüîç Strong Correlations (|r| > 0.5):")
    strong_correlations = []
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            corr_val = correlation_matrix.iloc[i, j]
            if abs(corr_val) > 0.5:
                strong_correlations.append((correlation_matrix.columns[i], correlation_matrix.columns[j], corr_val))
    
    if strong_correlations:
        for col1, col2, corr in strong_correlations:
            print(f"  ‚Ä¢ {col1} ‚Üî {col2}: {corr:.3f}")
    else:
        print("  No strong correlations found (|r| > 0.5)")
else:
    print("  ‚ö†Ô∏è Not enough numerical columns for correlation analysis")

# Êª°ÊÑèÂ∫¶ÂàÜÊûê
if 'satisfaction' in df.columns:
    print("\nüòä Customer Satisfaction Analysis:")
    
    # Êª°ÊÑèÂ∫¶‰∏éÈ£ûË°åË∑ùÁ¶ª
    if 'Flight Distance' in df.columns:
        plt.figure(figsize=(12, 6))
        sns.boxplot(data=df, x='satisfaction', y='Flight Distance')
        plt.title('Flight Distance vs. Satisfaction Level', fontsize=14, fontweight='bold')
        plt.xlabel('Satisfaction Level')
        plt.ylabel('Flight Distance (miles)')
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, 'flight_distance_vs_satisfaction.png'), 
                    dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()
    
    # Êª°ÊÑèÂ∫¶‰∏éÂπ¥ÈæÑ
    if 'Age' in df.columns:
        plt.figure(figsize=(12, 6))
        sns.kdeplot(data=df, x='Age', hue='satisfaction', fill=True, common_norm=False, alpha=0.7)
        plt.title('Age Distribution by Satisfaction Level', fontsize=14, fontweight='bold')
        plt.xlabel('Age (years)')
        plt.ylabel('Density')
        plt.legend(title='Satisfaction Level')
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, 'age_distribution_by_satisfaction.png'), 
                    dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()
    
    # Êª°ÊÑèÂ∫¶‰∏éÂÆ¢Êà∑Á±ªÂûã‰∫§ÂèâÂàÜÊûê
    if 'Customer Type' in df.columns:
        plt.figure(figsize=(12, 6))
        satisfaction_crosstab = pd.crosstab(df['Customer Type'], df['satisfaction'], normalize='index') * 100
        satisfaction_crosstab.plot(kind='bar', ax=plt.gca())
        plt.title('Satisfaction Rate by Customer Type', fontsize=14, fontweight='bold')
        plt.xlabel('Customer Type')
        plt.ylabel('Percentage (%)')
        plt.legend(title='Satisfaction Level', bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, 'satisfaction_by_customer_type.png'), 
                    dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()

# Âª∂ËØØÂàÜÊûê
delay_cols = ['Departure Delay in Minutes', 'Arrival Delay in Minutes']
if all(col in df.columns for col in delay_cols):
    print("\n‚úàÔ∏è Flight Delay Analysis:")
    
    # Ê£ÄÊü•Áº∫Â§±ÂÄºÊÉÖÂÜµ
    dept_missing = df[delay_cols[0]].isnull().sum()
    arr_missing = df[delay_cols[1]].isnull().sum()
    
    print(f"  Departure Delay missing values: {dept_missing}")
    print(f"  Arrival Delay missing values: {arr_missing}")
    
    # ÂàõÂª∫Ê≤°ÊúâÁº∫Â§±ÂÄºÁöÑÊï∞ÊçÆÂ≠êÈõÜÁî®‰∫éÁõ∏ÂÖ≥ÊÄßÂàÜÊûê
    delay_data = df[delay_cols].dropna()
    
    if len(delay_data) > 0:
        plt.figure(figsize=(12, 8))
        plt.scatter(delay_data[delay_cols[0]], delay_data[delay_cols[1]], alpha=0.6, s=20)
        plt.xlabel('Departure Delay (minutes)')
        plt.ylabel('Arrival Delay (minutes)')
        plt.title(f'Departure vs Arrival Delay\n(Based on {len(delay_data)} complete records, {arr_missing} missing values excluded)', 
                 fontsize=14, fontweight='bold')
        
        # Ê∑ªÂä†ÂõûÂΩíÁ∫ø
        if len(delay_data) > 1:
            z = np.polyfit(delay_data[delay_cols[0]], delay_data[delay_cols[1]], 1)
            p = np.poly1d(z)
            x_range = np.linspace(delay_data[delay_cols[0]].min(), delay_data[delay_cols[0]].max(), 100)
            plt.plot(x_range, p(x_range), "r-", alpha=0.8, linewidth=2, label='Trend Line')
            
            # ËÆ°ÁÆóÂπ∂ÊòæÁ§∫Áõ∏ÂÖ≥Á≥ªÊï∞
            correlation = delay_data[delay_cols[0]].corr(delay_data[delay_cols[1]])
            plt.text(0.05, 0.95, f'Correlation: {correlation:.3f}\nValid records: {len(delay_data)}\nMissing: {arr_missing}', 
                    transform=plt.gca().transAxes, 
                    bbox=dict(boxstyle="round,pad=0.3", facecolor='yellow', alpha=0.7),
                    fontsize=12, fontweight='bold')
            plt.legend()
        
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, 'departure_vs_arrival_delay.png'), 
                    dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()
        
        print(f"  ‚úÖ Delay analysis completed using {len(delay_data)} complete records")
    else:
        print("  ‚ö†Ô∏è No complete delay records available for analysis")

# ÂºÇÂ∏∏ÂÄºÊ£ÄÊµãÂíåÂèØËßÜÂåñ
print("\nüö® Outlier Detection Analysis:")
outlier_summary = {}
for col in numeric_cols:
    # Âè™ÂØπÈùûÁ©∫Êï∞ÊçÆËøõË°åÂºÇÂ∏∏ÂÄºÊ£ÄÊµã
    valid_data = df[col].dropna()
    
    if len(valid_data) > 0:
        Q1 = valid_data.quantile(0.25)
        Q3 = valid_data.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = valid_data[(valid_data < lower_bound) | (valid_data > upper_bound)]
        
        outlier_count = len(outliers)
        outlier_percentage = (outlier_count / len(valid_data)) * 100
        missing_count = df[col].isnull().sum()
        
        outlier_summary[col] = {
            'count': outlier_count, 
            'percentage': outlier_percentage,
            'missing': missing_count,
            'valid_data_count': len(valid_data)
        }
        
        # ÁâπÂà´Ê†áÊ≥®ÊúâÁº∫Â§±ÂÄºÁöÑÂàó
        missing_note = f" (Missing: {missing_count})" if missing_count > 0 else ""
        print(f"  ‚Ä¢ {col}: {outlier_count} outliers ({outlier_percentage:.1f}% of {len(valid_data)} valid records){missing_note}")
    else:
        print(f"  ‚Ä¢ {col}: No valid data available")
        outlier_summary[col] = {'count': 0, 'percentage': 0, 'missing': len(df), 'valid_data_count': 0}

# ÂºÇÂ∏∏ÂÄºÊ±áÊÄªÂèØËßÜÂåñ
if outlier_summary:
    plt.figure(figsize=(14, 8))
    cols = list(outlier_summary.keys())
    percentages = [outlier_summary[col]['percentage'] for col in cols]
    
    colors = plt.cm.Reds(np.linspace(0.3, 0.8, len(cols)))
    bars = plt.bar(range(len(cols)), percentages, color=colors)
    
    plt.xlabel('Attributes')
    plt.ylabel('Outlier Percentage (%)')
    plt.title('Outlier Detection Summary (% of Valid Data)', fontsize=14, fontweight='bold')
    plt.xticks(range(len(cols)), [col.replace(' ', '\n') for col in cols], rotation=45, ha='right')
    
    # Ê∑ªÂä†Êï∞ÂÄºÊ†áÁ≠æ
    for i, (bar, percentage) in enumerate(zip(bars, percentages)):
        missing_count = outlier_summary[cols[i]]['missing']
        label_text = f'{percentage:.1f}%'
        if missing_count > 0:
            label_text += f'\n(M:{missing_count})'
        plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + max(percentages)*0.01,
                label_text, ha='center', va='bottom', fontweight='bold', fontsize=8)
    
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'outlier_detection_summary.png'), 
                dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

print("\n" + "="*80)
print("TASK 1B: DATA PREPROCESSING")
print("="*80)

# ÂàõÂª∫È¢ÑÂ§ÑÁêÜÁªìÊûúÁöÑDataFrame
preprocessing_results = df.copy()

# Task 1B.1: ÂàÜÁÆ±ÊäÄÊúØ (Binning)
print("\nüì¶ Task 1B.1: Binning Techniques")
print("-" * 50)

delay_columns_for_binning = ['Departure Delay in Minutes', 'Arrival Delay in Minutes']

for col in delay_columns_for_binning:
    if col in df.columns:
        clean_col = col.replace(' ', '_').replace('-', '_').replace('(', '').replace(')', '')
        
        print(f"\nüîß Applying binning to: {col}")
        
        # Ê£ÄÊü•Áº∫Â§±ÂÄº
        missing_count = df[col].isnull().sum()
        total_count = len(df)
        valid_count = total_count - missing_count
        
        print(f"  üìä Data quality check:")
        print(f"    Total records: {total_count}")
        print(f"    Missing values: {missing_count} ({missing_count/total_count*100:.2f}%)")
        print(f"    Valid records: {valid_count} ({valid_count/total_count*100:.2f}%)")
        
        # Ëé∑ÂèñÈùûÁ©∫Êï∞ÊçÆËøõË°åÂàÜÊûê
        non_null_data = df[col].dropna()
        
        if len(non_null_data) == 0:
            print(f"  ‚ö†Ô∏è Warning: No valid data for {col}")
            continue
            
        print(f"  üìä Valid data range: {non_null_data.min():.1f} to {non_null_data.max():.1f}")
        
        # Á≠âÂÆΩÂàÜÁÆ± (Equal-width binning)
        n_bins = 5
        try:
            preprocessing_results[f'{clean_col}_equi_width'] = pd.cut(
                df[col], bins=n_bins, labels=[f'Bin_{i+1}' for i in range(n_bins)]
            )
            
            # ËÆ°ÁÆóÂàÜÁÆ±ËæπÁïå
            _, bin_edges_ew = pd.cut(non_null_data, bins=n_bins, retbins=True)
            
            print(f"  ‚úÖ Equal-width binning completed")
            print(f"    Bin edges: {[f'{edge:.1f}' for edge in bin_edges_ew]}")
            
            # ÁªüËÆ°ÂàÜÁÆ±ÁªìÊûú
            ew_counts = preprocessing_results[f'{clean_col}_equi_width'].value_counts()
            missing_in_binning = preprocessing_results[f'{clean_col}_equi_width'].isnull().sum()
            
            print(f"    Distribution:")
            for bin_name in [f'Bin_{i+1}' for i in range(n_bins)]:
                if bin_name in ew_counts.index:
                    count = ew_counts[bin_name]
                    percentage = (count / valid_count) * 100 if valid_count > 0 else 0
                    print(f"      {bin_name}: {count:,} samples ({percentage:.1f}% of valid data)")
            
            if missing_in_binning > 0:
                print(f"      Missing: {missing_in_binning:,} samples ({missing_count/total_count*100:.2f}% of total)")
            
        except Exception as e:
            print(f"  ‚ùå Equal-width binning failed: {e}")
        
        # Á≠âÊ∑±ÂàÜÁÆ± (Equal-depth binning)
        try:
            preprocessing_results[f'{clean_col}_equi_depth'] = pd.qcut(
                df[col], q=n_bins, labels=[f'Bin_{i+1}' for i in range(n_bins)], duplicates='drop'
            )
            
            print(f"  ‚úÖ Equal-depth binning completed")
            
            # ÁªüËÆ°ÂàÜÁÆ±ÁªìÊûú
            ed_counts = preprocessing_results[f'{clean_col}_equi_depth'].value_counts()
            
            print(f"    Distribution:")
            for bin_name in ed_counts.index:
                count = ed_counts[bin_name]
                percentage = (count / valid_count) * 100 if valid_count > 0 else 0
                print(f"      {bin_name}: {count:,} samples ({percentage:.1f}% of valid data)")
            
            if missing_count > 0:
                print(f"      Missing: {missing_count:,} samples ({missing_count/total_count*100:.2f}% of total)")
                
        except Exception as e:
            print(f"  ‚ùå Equal-depth binning failed: {e}")

        # ÂèØËßÜÂåñÂàÜÁÆ±ÁªìÊûú
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle(f'Binning Analysis for {col}\n(Total: {total_count:,}, Valid: {valid_count:,}, Missing: {missing_count:,})', 
                    fontsize=16, fontweight='bold')
        
        # ÂéüÂßãÊï∞ÊçÆÂàÜÂ∏É
        if len(non_null_data) > 0:
            sns.histplot(non_null_data, bins=30, ax=axes[0,0])
            axes[0,0].set_title(f'Original Distribution (Valid Data Only)\nValid: {len(non_null_data):,}, Missing: {missing_count:,}')
            axes[0,0].set_xlabel(col)
        
        # Á≠âÂÆΩÂàÜÁÆ±ÁªìÊûú
        if f'{clean_col}_equi_width' in preprocessing_results.columns:
            ew_counts = preprocessing_results[f'{clean_col}_equi_width'].value_counts()
            if len(ew_counts) > 0:
                bars = axes[0,1].bar(range(len(ew_counts)), ew_counts.values)
                axes[0,1].set_title('Equal-Width Binning')
                axes[0,1].set_xlabel('Bins')
                axes[0,1].set_ylabel('Count')
                axes[0,1].set_xticks(range(len(ew_counts)))
                axes[0,1].set_xticklabels(ew_counts.index, rotation=45)
                
                # Ê∑ªÂä†Êï∞ÂÄºÊ†áÁ≠æ
                for bar, count in zip(bars, ew_counts.values):
                    axes[0,1].text(bar.get_x() + bar.get_width()/2., bar.get_height() + max(ew_counts.values)*0.01,
                                  f'{count:,}', ha='center', va='bottom', fontweight='bold')
        
        # Á≠âÊ∑±ÂàÜÁÆ±ÁªìÊûú
        if f'{clean_col}_equi_depth' in preprocessing_results.columns:
            ed_counts = preprocessing_results[f'{clean_col}_equi_depth'].value_counts()
            if len(ed_counts) > 0:
                bars = axes[1,0].bar(range(len(ed_counts)), ed_counts.values)
                axes[1,0].set_title('Equal-Depth Binning')
                axes[1,0].set_xlabel('Bins')
                axes[1,0].set_ylabel('Count')
                axes[1,0].set_xticks(range(len(ed_counts)))
                axes[1,0].set_xticklabels(ed_counts.index, rotation=45)
                
                # Ê∑ªÂä†Êï∞ÂÄºÊ†áÁ≠æ
                for bar, count in zip(bars, ed_counts.values):
                    axes[1,0].text(bar.get_x() + bar.get_width()/2., bar.get_height() + max(ed_counts.values)*0.01,
                                  f'{count:,}', ha='center', va='bottom', fontweight='bold')
        
        # Áº∫Â§±ÂÄºÂ§ÑÁêÜÁ≠ñÁï•ÊØîËæÉ
        strategies = ['Keep as Missing', 'Most Frequent Bin', 'Separate Category']
        strategy_counts = [missing_count, missing_count, missing_count]
        
        if missing_count > 0:
            bars = axes[1,1].bar(strategies, strategy_counts, color=['lightcoral', 'lightblue', 'lightgreen'])
            axes[1,1].set_title(f'Missing Value Handling Strategies\n({missing_count} missing values)')
            axes[1,1].set_ylabel('Count')
            axes[1,1].tick_params(axis='x', rotation=45)
            
            # Ê∑ªÂä†Êï∞ÂÄºÊ†áÁ≠æ
            for bar in bars:
                height = bar.get_height()
                if height > 0:
                    axes[1,1].text(bar.get_x() + bar.get_width()/2., height + max(strategy_counts)*0.01,
                                  f'{int(height):,}', ha='center', va='bottom', fontweight='bold')
        else:
            axes[1,1].text(0.5, 0.5, 'No Missing Values', ha='center', va='center', transform=axes[1,1].transAxes, fontsize=12)
            axes[1,1].set_title('Missing Value Analysis')
        
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, f'{clean_col}_binning_analysis.png'), 
                    dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()

# Task 1B.2: Ê†áÂáÜÂåñÊäÄÊúØ (Normalization)
print("\nüìè Task 1B.2: Normalization Techniques")
print("-" * 50)

if 'Flight Distance' in df.columns:
    print("üîß Applying normalization to: Flight Distance")
    
    flight_distance = df['Flight Distance'].dropna()
    
    if len(flight_distance) > 0:
        # Min-Max Ê†áÂáÜÂåñ [0,1]
        min_val = flight_distance.min()
        max_val = flight_distance.max()
        preprocessing_results['Flight_Distance_min_max'] = (df['Flight Distance'] - min_val) / (max_val - min_val)
        
        # Z-score Ê†áÂáÜÂåñ
        mean_val = flight_distance.mean()
        std_val = flight_distance.std()
        preprocessing_results['Flight_Distance_z_score'] = (df['Flight Distance'] - mean_val) / std_val
        
        print(f"\nüìä Normalization Results:")
        print(f"  Original data:")
        print(f"    Min: {min_val:.1f}, Max: {max_val:.1f}")
        print(f"    Mean: {mean_val:.1f}, Std: {std_val:.1f}")
        print(f"  Min-Max normalized:")
        print(f"    Min: {preprocessing_results['Flight_Distance_min_max'].min():.4f}")
        print(f"    Max: {preprocessing_results['Flight_Distance_min_max'].max():.4f}")
        print(f"  Z-score normalized:")
        print(f"    Mean: {preprocessing_results['Flight_Distance_z_score'].mean():.4f}")
        print(f"    Std: {preprocessing_results['Flight_Distance_z_score'].std():.4f}")
        
        # ÂèØËßÜÂåñÊ†áÂáÜÂåñÁªìÊûú
        fig, axes = plt.subplots(1, 3, figsize=(18, 6))
        fig.suptitle('Flight Distance Normalization Analysis', fontsize=16, fontweight='bold')
        
        # ÂéüÂßãÊï∞ÊçÆ
        sns.histplot(df['Flight Distance'].dropna(), kde=True, ax=axes[0])
        axes[0].set_title('Original Distribution')
        axes[0].set_xlabel('Flight Distance (miles)')
        
        # Min-MaxÊ†áÂáÜÂåñ
        sns.histplot(preprocessing_results['Flight_Distance_min_max'].dropna(), kde=True, ax=axes[1])
        axes[1].set_title('Min-Max Normalized [0,1]')
        axes[1].set_xlabel('Normalized Flight Distance')
        
        # Z-scoreÊ†áÂáÜÂåñ
        sns.histplot(preprocessing_results['Flight_Distance_z_score'].dropna(), kde=True, ax=axes[2])
        axes[2].set_title('Z-Score Normalized')
        axes[2].set_xlabel('Z-Score Flight Distance')
        
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, 'flight_distance_normalization.png'), 
                    dpi=300, bbox_inches='tight')
        plt.show()
        plt.close()

# Task 1B.3:  Discretization
print("\n Task 1B.3: Age Discretization")
print("-" * 50)

if 'Age' in df.columns:
    print(" Discretizing age into life stage categories...")
    
    def categorize_age(age):
        if pd.isna(age):
            return np.nan
        elif age <= 21:
            return 'Young'
        elif 22 <= age <= 34:
            return 'Early Adulthood'
        elif 35 <= age <= 44:
            return 'Early Middle Age'
        elif 45 <= age <= 64:
            return 'Late Middle Age'
        else:
            return 'Late Adulthood'
    
    preprocessing_results['Age_Category'] = df['Age'].apply(categorize_age)
    
    # ÁªüËÆ°Âπ¥ÈæÑÁ±ªÂà´È¢ëÁéá
    age_freq = preprocessing_results['Age_Category'].value_counts()
    total_valid = age_freq.sum()
    
    print(f"\n Age Category Frequencies:")
    category_order = ['Young', 'Early Adulthood', 'Early Middle Age', 'Late Middle Age', 'Late Adulthood']
    for category in category_order:
        if category in age_freq.index:
            count = age_freq[category]
            percentage = (count / total_valid) * 100
            print(f"  ‚Ä¢ {category}: {count:,} ({percentage:.1f}%)")
    
    # Âπ¥ÈæÑÂàÜÁ±ªÂèØËßÜÂåñ
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle('Age Analysis and Categorization', fontsize=16, fontweight='bold')
    
    # ÂéüÂßãÂπ¥ÈæÑÂàÜÂ∏É
    sns.histplot(df['Age'].dropna(), bins=30, kde=True, ax=axes[0])
    axes[0].set_title('Original Age Distribution')
    axes[0].set_xlabel('Age (years)')
    axes[0].set_ylabel('Count')
    
    # Âπ¥ÈæÑÁ±ªÂà´ÂàÜÂ∏É
    age_freq_ordered = age_freq.reindex(category_order).fillna(0)
    colors = plt.cm.viridis(np.linspace(0, 1, len(age_freq_ordered)))
    bars = axes[1].bar(range(len(age_freq_ordered)), age_freq_ordered.values, color=colors)
    axes[1].set_title('Age Category Distribution')
    axes[1].set_xlabel('Age Category')
    axes[1].set_ylabel('Count')
    axes[1].set_xticks(range(len(age_freq_ordered)))
    axes[1].set_xticklabels([cat.replace(' ', '\n') for cat in age_freq_ordered.index], rotation=45)
    
    # Ê∑ªÂä†ËÆ°Êï∞Ê†áÁ≠æ
    for i, (bar, count) in enumerate(zip(bars, age_freq_ordered.values)):
        if count > 0:
            axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + total_valid*0.01,
                        f'{int(count):,}', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'age_discretization_analysis.png'), 
                dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

# Task 1B.4: (Satisfaction Binarization)
print("\n Task 1B.4: Satisfaction Binarization")
print("-" * 50)

if 'satisfaction' in df.columns:
    print(" Converting satisfaction to binary values...")
    
    # Êü•ÁúãÂéüÂßãÊª°ÊÑèÂ∫¶ÂàÜÂ∏É
    original_satisfaction = df['satisfaction'].value_counts()
    print(f"\n Original Satisfaction Distribution:")
    for level, count in original_satisfaction.items():
        percentage = (count / len(df)) * 100
        print(f"  ‚Ä¢ {level}: {count:,} ({percentage:.1f}%)")
    
    # Â∞ÜÊª°ÊÑèÂ∫¶ËΩ¨Êç¢‰∏∫‰∫åËøõÂà∂
    satisfaction_mapping = {
        'satisfied': 1,
        'neutral or dissatisfied': 0
    }
    
    preprocessing_results['satisfaction_binary'] = df['satisfaction'].map(satisfaction_mapping)
    
    # ÁªüËÆ°‰∫åÂÄºÂåñÁªìÊûú
    binary_freq = preprocessing_results['satisfaction_binary'].value_counts().sort_index()
    total_valid = binary_freq.sum()
    
    print(f"\n Binarized Satisfaction Results:")
    print(f"  ‚Ä¢ Not Satisfied (0): {binary_freq.get(0, 0):,} ({binary_freq.get(0, 0)/total_valid*100:.1f}%)")
    print(f"  ‚Ä¢ Satisfied (1): {binary_freq.get(1, 0):,} ({binary_freq.get(1, 0)/total_valid*100:.1f}%)")
    
    # Êª°ÊÑèÂ∫¶‰∫åÂÄºÂåñÂèØËßÜÂåñ
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    fig.suptitle('Satisfaction Binarization Analysis', fontsize=16, fontweight='bold')
    
    # ÂéüÂßãÊª°ÊÑèÂ∫¶ÂàÜÂ∏É
    original_satisfaction.plot(kind='bar', ax=axes[0], color=['lightcoral', 'lightgreen'])
    axes[0].set_title('Original Satisfaction Distribution')
    axes[0].set_xlabel('Satisfaction Level')
    axes[0].set_ylabel('Count')
    axes[0].tick_params(axis='x', rotation=45)
    
    # ‰∫åÂÄºÂåñÁªìÊûú
    binary_labels = ['Not Satisfied (0)', 'Satisfied (1)']
    colors = ['lightcoral', 'lightgreen']
    bars = axes[1].bar(binary_labels, [binary_freq.get(0, 0), binary_freq.get(1, 0)], color=colors)
    axes[1].set_title('Binarized Satisfaction Distribution')
    axes[1].set_xlabel('Binary Satisfaction')
    axes[1].set_ylabel('Count')
    
    # Ê∑ªÂä†ÁôæÂàÜÊØîÊ†áÁ≠æ
    for bar, count in zip(bars, [binary_freq.get(0, 0), binary_freq.get(1, 0)]):
        percentage = (count / total_valid) * 100
        axes[1].text(bar.get_x() + bar.get_width()/2, bar.get_height() + total_valid*0.01,
                    f'{count:,}\n({percentage:.1f}%)', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.savefig(os.path.join(output_folder, 'satisfaction_binarization_analysis.png'), 
                dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()

# ‰øùÂ≠òÊâÄÊúâÈ¢ÑÂ§ÑÁêÜÁªìÊûúÂà∞ExcelÊñá‰ª∂
print(f"\nüíæ Saving all results to Excel file...")
print(f"üìÅ Excel file location: {excel_file_path}")

try:
    with pd.ExcelWriter(excel_file_path, engine='openpyxl') as writer:
        # 1A - ÂéüÂßãÊï∞ÊçÆÂíåÂü∫Êú¨ÂàÜÊûê
        df.to_excel(writer, sheet_name='1A_Original_Data', index=False)
        print("  ‚úÖ Sheet '1A_Original_Data' saved")
        
        # Êï∞ÂÄºÂ±ûÊÄßÊ±áÊÄªÁªüËÆ°
        numeric_summary.to_excel(writer, sheet_name='1A_Numeric_Summary')
        print("  ‚úÖ Sheet '1A_Numeric_Summary' saved")
        
        # Áõ∏ÂÖ≥ÊÄßÁü©Èòµ
        if 'correlation_matrix' in locals():
            correlation_matrix.to_excel(writer, sheet_name='1A_Correlation_Matrix')
            print("  ‚úÖ Sheet '1A_Correlation_Matrix' saved")
        
        # ÂàÜÁ±ªÂ±ûÊÄßÈ¢ëÁéáÁªüËÆ°
        categorical_summary = pd.DataFrame()
        for col in categorical_cols:
            freq_data = df[col].value_counts().reset_index()
            freq_data.columns = [col, 'Count']
            freq_data['Percentage'] = (freq_data['Count'] / len(df)) * 100
            freq_data['Attribute'] = col
            categorical_summary = pd.concat([categorical_summary, freq_data], ignore_index=True)
        categorical_summary.to_excel(writer, sheet_name='1A_Categorical_Summary', index=False)
        print("  ‚úÖ Sheet '1A_Categorical_Summary' saved")
        
        # Áº∫Â§±ÂÄºÂàÜÊûê
        missing_analysis = pd.DataFrame({
            'Column': df.columns,
            'Missing_Count': [df[col].isnull().sum() for col in df.columns],
            'Missing_Percentage': [(df[col].isnull().sum() / len(df)) * 100 for col in df.columns],
            'Data_Type': [str(df[col].dtype) for col in df.columns]
        })
        missing_analysis.to_excel(writer, sheet_name='1A_Missing_Values_Analysis', index=False)
        print("  ‚úÖ Sheet '1A_Missing_Values_Analysis' saved")
        
        # 1B - ÂÆåÊï¥ÁöÑÈ¢ÑÂ§ÑÁêÜÁªìÊûú
        preprocessing_results.to_excel(writer, sheet_name='1B_All_Preprocessing', index=False)
        print("  ‚úÖ Sheet '1B_All_Preprocessing' saved")
        
        # ÂàÜÁÆ±ÁªìÊûúÂçïÁã¨‰øùÂ≠ò
        binning_data = pd.DataFrame()
        for col in delay_columns_for_binning:
            if col in df.columns:
                clean_col = col.replace(' ', '_').replace('-', '_').replace('(', '').replace(')', '')
                temp_data = pd.DataFrame({
                    f'Original_{col.replace(" ", "_")}': df[col]
                })
                if f'{clean_col}_equi_width' in preprocessing_results.columns:
                    temp_data[f'{col.replace(" ", "_")}_Equi_Width'] = preprocessing_results[f'{clean_col}_equi_width']
                if f'{clean_col}_equi_depth' in preprocessing_results.columns:
                    temp_data[f'{col.replace(" ", "_")}_Equi_Depth'] = preprocessing_results[f'{clean_col}_equi_depth']
                
                if binning_data.empty:
                    binning_data = temp_data
                else:
                    binning_data = pd.concat([binning_data, temp_data], axis=1)
        
        if not binning_data.empty:
            binning_data.to_excel(writer, sheet_name='1B_Binning_Results', index=False)
            print("  ‚úÖ Sheet '1B_Binning_Results' saved")
        
        # Ê†áÂáÜÂåñÁªìÊûú
        if 'Flight Distance' in df.columns and 'Flight_Distance_min_max' in preprocessing_results.columns:
            normalization_data = pd.DataFrame({
                'Flight_Distance_Original': df['Flight Distance'],
                'Flight_Distance_Min_Max': preprocessing_results['Flight_Distance_min_max'],
                'Flight_Distance_Z_Score': preprocessing_results['Flight_Distance_z_score']
            })
            normalization_data.to_excel(writer, sheet_name='1B_Normalization', index=False)
            print("  ‚úÖ Sheet '1B_Normalization' saved")
        
        # Âπ¥ÈæÑÁ¶ªÊï£ÂåñÁªìÊûú
        if 'Age' in df.columns and 'Age_Category' in preprocessing_results.columns:
            age_data = pd.DataFrame({
                'Age_Original': df['Age'],
                'Age_Category': preprocessing_results['Age_Category']
            })
            age_data.to_excel(writer, sheet_name='1B_Age_Discretization', index=False)
            print("  ‚úÖ Sheet '1B_Age_Discretization' saved")
        
        # Êª°ÊÑèÂ∫¶‰∫åÂÄºÂåñÁªìÊûú
        if 'satisfaction' in df.columns and 'satisfaction_binary' in preprocessing_results.columns:
            satisfaction_data = pd.DataFrame({
                'Satisfaction_Original': df['satisfaction'],
                'Satisfaction_Binary': preprocessing_results['satisfaction_binary']
            })
            satisfaction_data.to_excel(writer, sheet_name='1B_Satisfaction_Binary', index=False)
            print("  ‚úÖ Sheet '1B_Satisfaction_Binary' saved")
        
        # ÂºÇÂ∏∏ÂÄºÊ£ÄÊµãÁªìÊûú
        if outlier_summary:
            outlier_df = pd.DataFrame([
                {
                    'Attribute': col,
                    'Outlier_Count': info['count'],
                    'Outlier_Percentage': info['percentage'],
                    'Missing_Count': info['missing'],
                    'Valid_Data_Count': info['valid_data_count']
                }
                for col, info in outlier_summary.items()
            ])
            outlier_df.to_excel(writer, sheet_name='1A_Outlier_Analysis', index=False)
            print("  ‚úÖ Sheet '1A_Outlier_Analysis' saved")

    print(f"‚úÖ Excel file successfully saved!")

except Exception as e:
    print(f"‚ùå Error saving Excel file: {e}")

print("\n" + "="*80)
print("TASK 1C: SUMMARY OF FINDINGS")
print("="*80)

print(f"\nüéØ EXECUTIVE SUMMARY FOR HEAD OF ANALYTICS UNIT")
print("=" * 60)

# Êï∞ÊçÆÈõÜÊ¶ÇËßà
print(f"\nüìä DATASET OVERVIEW:")
print(f"  ‚Ä¢ Total Records: {len(df):,}")
print(f"  ‚Ä¢ Total Attributes: {len(df.columns)}")
print(f"  ‚Ä¢ Numerical Attributes: {len(numeric_cols)}")
print(f"  ‚Ä¢ Categorical Attributes: {len(categorical_cols)}")

# ËÆ°ÁÆóÊÄª‰ΩìÊï∞ÊçÆÂÆåÊï¥ÊÄß
total_cells = len(df) * len(df.columns)
missing_cells = df.isnull().sum().sum()
completeness = ((total_cells - missing_cells) / total_cells) * 100

print(f"  ‚Ä¢ Data Completeness: {completeness:.2f}%")
if missing_cells > 0:
    print(f"  ‚Ä¢ Total Missing Values: {missing_cells}")
    print(f"  ‚Ä¢ Columns with Missing Data:")
    for col, count in df.isnull().sum()[df.isnull().sum() > 0].items():
        print(f"    - {col}: {count} ({count/len(df)*100:.2f}%)")
else:
    print(f"  ‚Ä¢ ‚úÖ No missing values")

# ÂÆ¢Êà∑Êª°ÊÑèÂ∫¶ÂÖ≥ÈîÆÊ¥ûÂØü
if 'satisfaction' in df.columns:
    satisfaction_dist = df['satisfaction'].value_counts()
    satisfied_rate = (satisfaction_dist.get('satisfied', 0) / len(df)) * 100
    print(f"\nüòä CUSTOMER SATISFACTION INSIGHTS:")
    print(f"  ‚Ä¢ Overall Satisfaction Rate: {satisfied_rate:.1f}%")
    
    for level, count in satisfaction_dist.items():
        percentage = (count / len(df)) * 100
        print(f"  ‚Ä¢ {level.title()}: {count:,} passengers ({percentage:.1f}%)")

# ÂºÇÂ∏∏ÂÄºÂíåÊï∞ÊçÆË¥®ÈáèÈóÆÈ¢ò
print(f"\nüö® DATA QUALITY ISSUES:")
if outlier_summary:
    high_outlier_cols = [col for col, info in outlier_summary.items() if info['percentage'] > 5]
    if high_outlier_cols:
        print(f"  ‚Ä¢ High Outlier Attributes (>5%):")
        for col in high_outlier_cols[:5]:
            print(f"    - {col}: {outlier_summary[col]['percentage']:.1f}% outliers")
    else:
        print(f"  ‚Ä¢ ‚úÖ Outlier levels are manageable (<5% per attribute)")
else:
    print(f"  ‚Ä¢ No outlier analysis available")

# Âº∫Áõ∏ÂÖ≥ÊÄßÂèëÁé∞
print(f"\nüîó KEY RELATIONSHIPS IDENTIFIED:")
if 'strong_correlations' in locals() and strong_correlations:
    print(f"  ‚Ä¢ Strong Correlations Found ({len(strong_correlations)}):")
    for col1, col2, corr in strong_correlations[:3]:
        correlation_strength = "Very Strong" if abs(corr) > 0.8 else "Strong"
        print(f"    - {col1} ‚Üî {col2}: {corr:.3f} ({correlation_strength})")
else:
    print(f"  ‚Ä¢ No strong correlations detected or analysis not completed")

# Âª∫ËÆÆÁöÑÂêéÁª≠Ê≠•È™§
print(f"\nüéØ RECOMMENDED NEXT STEPS:")
print(f"  1. Handle missing values in Arrival Delay using appropriate imputation")
print(f"  2. Implement satisfaction prediction model using preprocessed features")
print(f"  3. Focus on delay management strategies")
print(f"  4. Develop targeted satisfaction improvement programs by customer segment")
print(f"  5. Consider additional feature engineering for flight distance and delays")

# ÊäÄÊúØË¥®ÈáèËØÑ‰º∞
print(f"\nüîß TECHNICAL PREPROCESSING COMPLETED:")
print(f"  ‚úÖ Missing Value Analysis: Comprehensive analysis completed")
print(f"  ‚úÖ Binning: Equal-width and equal-depth binning applied to delay variables")
print(f"  ‚úÖ Normalization: Min-Max and Z-score normalization for flight distance")
print(f"  ‚úÖ Discretization: Age categorized into 5 life-stage groups")
print(f"  ‚úÖ Binarization: Satisfaction converted to binary classification target")

# ÁîüÊàêÊúÄÁªà‰ª™Ë°®ÊùøÊÄªËßà
print(f"\nüìà Generating final executive dashboard...")
fig = plt.figure(figsize=(20, 12))
gs = fig.add_gridspec(3, 4, hspace=0.3, wspace=0.3)

# 1. Êª°ÊÑèÂ∫¶ÂàÜÂ∏É
ax1 = fig.add_subplot(gs[0, 0])
if 'satisfaction' in df.columns:
    satisfaction_counts = df['satisfaction'].value_counts()
    colors = ['lightcoral', 'lightgreen']
    ax1.pie(satisfaction_counts.values, labels=satisfaction_counts.index, autopct='%1.1f%%', colors=colors)
    ax1.set_title('Customer Satisfaction\nDistribution', fontweight='bold')

# 2. Âπ¥ÈæÑÂàÜÂ∏É
ax2 = fig.add_subplot(gs[0, 1])
if 'Age' in df.columns and 'Age_Category' in preprocessing_results.columns:
    age_categories = preprocessing_results['Age_Category'].value_counts()
    category_order = ['Young', 'Early Adulthood', 'Early Middle Age', 'Late Middle Age', 'Late Adulthood']
    age_ordered = age_categories.reindex(category_order).fillna(0)
    ax2.bar(range(len(age_ordered)), age_ordered.values, color=plt.cm.viridis(np.linspace(0, 1, len(age_ordered))))
    ax2.set_title('Age Category\nDistribution', fontweight='bold')
    ax2.set_xticks(range(len(age_ordered)))
    ax2.set_xticklabels([cat.replace(' ', '\n') for cat in age_ordered.index], rotation=45, fontsize=8)

# 3. ÂºÇÂ∏∏ÂÄºÊ±áÊÄª
ax3 = fig.add_subplot(gs[0, 2])
if outlier_summary:
    outlier_cols = list(outlier_summary.keys())[:6]
    outlier_percentages = [outlier_summary[col]['percentage'] for col in outlier_cols]
    
    bars = ax3.bar(range(len(outlier_cols)), outlier_percentages, color=plt.cm.Reds(np.linspace(0.3, 0.8, len(outlier_cols))))
    ax3.set_title('Outlier Detection\nSummary', fontweight='bold')
    ax3.set_ylabel('Outlier %')
    ax3.set_xticks(range(len(outlier_cols)))
    ax3.set_xticklabels([col.replace(' ', '\n') for col in outlier_cols], rotation=45, fontsize=8)

# 4. Áº∫Â§±ÂÄºÂàÜÊûê
ax4 = fig.add_subplot(gs[0, 3])
missing_counts = [df[col].isnull().sum() for col in df.columns]
cols_with_missing = [i for i, count in enumerate(missing_counts) if count > 0]
if cols_with_missing:
    missing_cols = [df.columns[i] for i in cols_with_missing]
    missing_values_list = [missing_counts[i] for i in cols_with_missing]
    ax4.bar(range(len(missing_cols)), missing_values_list, color='orange')
    ax4.set_title('Missing Values\nby Column', fontweight='bold')
    ax4.set_ylabel('Missing Count')
    ax4.set_xticks(range(len(missing_cols)))
    ax4.set_xticklabels([col.replace(' ', '\n') for col in missing_cols], rotation=45, fontsize=8)
else:
    ax4.text(0.5, 0.5, 'No Missing\nValues', ha='center', va='center', transform=ax4.transAxes, fontsize=12)
    ax4.set_title('Missing Values\nAnalysis', fontweight='bold')

# 5. Áõ∏ÂÖ≥ÊÄßÁÉ≠ÂõæÔºàÁÆÄÂåñÁâàÔºâ
ax5 = fig.add_subplot(gs[1, :2])
if 'correlation_matrix' in locals():
    if len(correlation_matrix.columns) > 6:
        correlation_subset = correlation_matrix.iloc[:6, :6]
    else:
        correlation_subset = correlation_matrix
    sns.heatmap(correlation_subset, annot=True, cmap='RdBu_r', center=0, ax=ax5, fmt='.2f', cbar=False)
    ax5.set_title('Correlation Matrix (Key Variables)', fontweight='bold')
else:
    ax5.text(0.5, 0.5, 'Correlation Matrix\nNot Available', ha='center', va='center', transform=ax5.transAxes, fontsize=12)
    ax5.set_title('Correlation Analysis', fontweight='bold')

# 6. Âª∂ËØØÂÖ≥Á≥ªÂõæ
ax6 = fig.add_subplot(gs[1, 2:])
if all(col in df.columns for col in delay_cols):
    delay_sample = df[delay_cols].dropna()
    if len(delay_sample) > 0:
        sample_size = min(1000, len(delay_sample))
        delay_sample = delay_sample.sample(sample_size)
        ax6.scatter(delay_sample[delay_cols[0]], delay_sample[delay_cols[1]], alpha=0.6, s=10)
        ax6.set_xlabel('Departure Delay (min)')
        ax6.set_ylabel('Arrival Delay (min)')
        ax6.set_title(f'Flight Delays Relationship\n({len(delay_sample)} valid records)', fontweight='bold')

# 7. È¢ÑÂ§ÑÁêÜÊäÄÊúØÊ±áÊÄª
ax7 = fig.add_subplot(gs[2, :])
preprocessing_techniques = ['Missing Value\nAnalysis', 'Binning\n(Equal-width & Equal-depth)', 
                           'Normalization\n(Min-Max & Z-score)', 'Discretization\n(Age Categories)', 
                           'Binarization\n(Satisfaction)']
technique_status = ['‚úÖ Completed', '‚úÖ Completed', '‚úÖ Completed', '‚úÖ Completed', '‚úÖ Completed']

ax7.barh(range(len(preprocessing_techniques)), [1]*len(preprocessing_techniques), color='lightgreen')
ax7.set_yticks(range(len(preprocessing_techniques)))
ax7.set_yticklabels(preprocessing_techniques)
ax7.set_xlabel('Completion Status')
ax7.set_title('Data Preprocessing Tasks Completed', fontweight='bold')
ax7.set_xlim(0, 1.2)

for i, status in enumerate(technique_status):
    ax7.text(0.5, i, status, ha='center', va='center', fontweight='bold')

plt.suptitle('AIRLINE PASSENGER SATISFACTION ANALYSIS - EXECUTIVE DASHBOARD\nStudent ID: 14645473', 
             fontsize=20, fontweight='bold', y=0.98)
plt.savefig(os.path.join(output_folder, 'executive_dashboard_summary.png'), 
            dpi=300, bbox_inches='tight')
plt.show()
plt.close()

print("\n" + "="*80)
print("‚úÖ ANALYSIS COMPLETE!")
print("="*80)

print(f"\nüìÅ ALL FILES SAVED TO:")
print(f"  üìä Analysis Results Folder: {output_folder}")
print(f"  üìã Excel Workbook: {excel_file_path}")

print(f"\nüìà GENERATED VISUALIZATIONS:")
visualization_files = [
    "Individual attribute detailed analysis charts (with missing value info)",
    "Categorical attribute frequency distribution charts",
    "Correlation matrix heatmap", 
    "Satisfaction vs. key variables analysis",
    "Flight delay relationship analysis (accounting for missing values)",
    "Outlier detection summary (with missing value annotations)",
    "Binning analysis charts (with missing value handling strategies)",
    "Normalization comparison charts", 
    "Age discretization analysis",
    "Satisfaction binarization analysis",
    "Executive dashboard summary"
]

for i, viz in enumerate(visualization_files, 1):
    print(f"  {i:2d}. {viz}")

print(f"\nüéØ READY FOR REPORT WRITING!")
print(f"‚úÖ All missing values properly identified and handled")
print(f"‚úÖ All preprocessing tasks completed with quality checks")
print(f"‚úÖ Comprehensive visualizations generated with data quality annotations")
print(f"‚úÖ Excel workbook contains all results for easy reference")

print("\n" + "="*80)
```
